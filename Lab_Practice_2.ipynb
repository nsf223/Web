```python
import numpy as np
%pylab inline
import matplotlib.pyplot as plt

def control(t, y):
    ux = 0
    uy = 0
    uz = 0
    return [ux, uy, uz]

def simulate(dt, x, u, m=1, g=9.8):
    # Control input affects the acceleration
    dx = np.array([x[3], x[4], x[5], u[0]/m, u[1]/m, u[2]/m])

    # Include the gravity in the acceleration in z
    dx[5] += -g

    # Euler integration
    x += dt * dx

    # Simulate the floor (Plastic collision)
    if x[2] < 0:
        x[2] = 0  # z = 0
        x[5] = 0  # vz = 0

    return x

def sense(x):
  return x

tf = 10
dt = 0.1
time = np.linspace(0, tf, 100)
x = np.array([0., 0., 10., 0.,0.,0.])
x_log = [np.copy(x)]

for t in time:
  y = sense(x)
  u = control(t, y)
  x = simulate(dt, x, u)
  x_log.append(np.copy(x))
x_log = np.array(x_log)

fig, ax = plt.subplots()
ax.plot(time, x_log[:-1,2])
ax.set(title="Change in altitude with respect to time", xlabel="Time", ylabel="Height")
ax.grid()
```

```python
import numpy as np
%pylab inline
import matplotlib.pyplot as plt

ta=0

def control(t, y):
    ux = 0
    uy = 0
    uz = t*(15-y[2])+(t/6)*(0-y[5])+9.8
    return [ux, uy, uz]

def simulate(dt, x, u, m=0.8, g=9.8):

    # Control input affects the acceleration
    dx = np.array([x[3], x[4], x[5], u[0]/m, u[1]/m, u[2]/m])

    # Include the gravity in the acceleration in z
    dx[5] += -g

    # Euler integration
    x += dt * dx

    # Simulate the floor (Plastic collision)
    if x[2] < 0:
        x[2] = 0  # z = 0
        x[5] = 0  # vz = 0

    return x

def sense(x):
  return x

tf = 60
dt = 0.1
time = np.linspace(0, tf, 100)
x = np.array([0., 0., 10., 0.,0.,0.])
x_log = [np.copy(x)]

for t in time:
  y = sense(x)
  u = control(t, y)
  x = simulate(dt, x, u)
  if x[2] > 15:
    ta += 1
  x_log.append(np.copy(x))
x_log = np.array(x_log)

fig, ax = plt.subplots()
ax.plot(time, x_log[:-1,2])
ax.set(title="Change in altitude with respect to time", xlabel="Time", ylabel="Height")
ax.grid()
```
